```python
# Cоздадим функцию для построения бар диаграмм для двух разных категориальных признаков 
""" Функция для создания диаграмм плотности рассеяния
Принимает имя датафрейма, названия двух разных столбцов с категориальными признаками. Выдает графики
"""
def plot_grouped_bar(df, categorical_column1, categorical_column2): # Вычисляем количество для каждой кимбинации категорий
    data = df.groupby(categorical_column1)[categorical_column2].value_counts().unstack()
    # Создаем общую бар-диаграмму
    plt.figure(figsize=(10, 8))
    data.plot(kind='bar', stacked=False)
    plt.xlabel(categorical_column1, size=14)
    plt.ylabel('Count', size=14)
    plt.title(f'Grouped Bar Diagram of {categorical_column2} by {categorical_column1}', size=16) 
    plt.xticks(rotation=45, ha='right')
    plt.grid() 
    plt.tight_layout() 
    plt.show()
```

```python


max_fields = 39  # Set the maximum number of fields

# Function to split the string and pad with 'none'
def split_and_pad(s):
    fields = s.split('|')
    # Pad the list if it has less than max_fields elements
    padded_fields = fields[:max_fields] + ['unknown'] * max(0, max_fields - len(fields))
    assert len(padded_fields) == max_fields, f"Expected {max_fields} fields, got {len(padded_fields)}"
    return padded_fields

# Apply the function to the column and create a new DataFrame
new_columns = df_train_raw['category_list'].apply(split_and_pad).tolist()
new_df = pd.DataFrame(new_columns, columns=[f'Field_{i+1}' for i in range(max_fields)])

# Concatenate the new columns to the original DataFrame
#df = pd.concat([df_train_raw, new_df], axis=1)

# Optionally, drop the original 'category_list' column if no longer needed
# df = df.drop('category_list', axis=1)

print(new_df.columns)
```

```python
def remove_duplicates_and_keep_dict_structure(industry_terms):
    seen = set()  # Set to keep track of seen terms
    new_dict = {}  # New dictionary to store unique values
    for category, terms in industry_terms.items():
        new_terms = []
        for term in terms:
            if term not in seen:
                seen.add(term)
                new_terms.append(term)
        if new_terms:  # Only add to new_dict if there are new terms
            new_dict[category] = new_terms
    return new_dict


unique_industry_terms = remove_duplicates_and_keep_dict_structure(industry_terms)
unique_industry_terms
```
